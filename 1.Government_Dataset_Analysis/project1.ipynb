```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("CRIME_REVIEW_FOR_THE_MONTH_OF_SEPTEMBER_2024_0.csv")
df
```

```python
df.info
```

```python
numeric_columns = df.columns[4:]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
sns.set_style("whitegrid")
```

```python
# 1. Total Crimes by Minor Head
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Minor Heads", y="During the current year upto the end of month under review")
plt.title("Total Crimes by Minor Head")
plt.xticks(rotation=90)
plt.show()
```

```python
# 2. Crime Comparison (Current vs Previous Year)
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Minor Heads", y="During the corresponding month of previous year")
plt.title("Crime Comparison (Current vs Previous Year)")
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Crime Trends Over Months
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Minor Heads", y="During the previous month")
plt.title("Crime Trends Over Months")
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Crime Trends in Current Month
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Minor Heads", y="During the current month")
plt.title("Crime Trends in Current Month")
plt.xticks(rotation=90)
plt.show()
```

```python
# 5. Distribution of Current Year Crimes
plt.figure(figsize=(10, 5))
sns.histplot(df["During the current year upto the end of month under review"], bins=10, kde=True)
plt.title("Distribution of Current Year Crimes")
plt.show()
```

```python
# 6. Distribution of Previous Year Crimes
plt.figure(figsize=(10, 5))
sns.histplot(df["During the corresponding month of previous year"], bins=10, kde=True)
plt.title("Distribution of Previous Year Crimes")
plt.show()
```

```python
# 7. Boxplot of Current Year Crimes
plt.figure(figsize=(5, 5))
sns.boxplot(y=df["During the current year upto the end of month under review"])
plt.title("Boxplot of Current Year Crimes")
plt.show()
```

```python
# 8. Boxplot of Previous Year Crimes
plt.figure(figsize=(5, 5))
sns.boxplot(y=df["During the corresponding month of previous year"])
plt.title("Boxplot of Previous Year Crimes")
plt.show()
```

```python
# 9. Crime Counts by Major Heads
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Major Heads", y="During the current year upto the end of month under review")
plt.title("Crime Counts by Major Heads")
plt.xticks(rotation=90)
plt.show()
```

```python
# 10. Crime Severity Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["During the current month"], bins=10, kde=True)
plt.title("Crime Severity Distribution")
plt.show()
```

```python
# 11. Crime Trend: Current vs Previous Month
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="During the previous month", y="During the current month")
plt.title("Crime Trend: Current vs Previous Month")
plt.show()
```

```python
# 12. Crime Trend: Current Year vs Previous Year
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="During the corresponding month of previous year", y="During the current year upto the end of month under review")
plt.title("Crime Trend: Current Year vs Previous Year")
plt.show()
```

```python
# 13. Correlation Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 14. Crime Count by Major Head
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Major Heads", y="During the current year upto the end of month under review")
plt.title("Crime Count by Major Head")
plt.xticks(rotation=90)
plt.show()
```

```python
# 15. Crime Count by Minor Head
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Minor Heads", y="During the current year upto the end of month under review")
plt.title("Crime Count by Minor Head")
plt.xticks(rotation=90)
plt.show()
```

```python
# 16. Top 10 Most Frequent Crimes
plt.figure(figsize=(10, 5))
top10 = df.groupby("Minor Heads")["During the current year upto the end of month under review"].sum().nlargest(10)
sns.barplot(x=top10.index, y=top10.values)
plt.title("Top 10 Most Frequent Crimes")
plt.xticks(rotation=90)
plt.show()
```

```python
# 17. Pie Chart of Crimes in Current Month
plt.figure(figsize=(7, 7))
df["During the current month"].dropna().plot(kind="pie", autopct="%1.1f%%")
plt.title("Pie Chart of Crimes in Current Month")
plt.ylabel("")
plt.show()
```

```python
# 18. Pie Chart of Crimes in Previous Year
plt.figure(figsize=(7, 7))
df["During the corresponding month of previous year"].dropna().plot(kind="pie", autopct="%1.1f%%")
plt.title("Pie Chart of Crimes in Previous Year")
plt.ylabel("")
plt.show()
```

```python
# 19. Boxplot: Current Month vs Previous Month
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[["During the previous month", "During the current month"]])
plt.title("Boxplot: Current Month vs Previous Month")
plt.show()
```

```python
# 20. Distribution of Crime Numbers
plt.figure(figsize=(10, 5))
sns.histplot(df["During the current year upto the end of month under review"], bins=10, kde=True)
plt.title("Distribution of Crime Numbers")
plt.show()
```

#Key Observations from the Crime Data:
Dataset Overview:

The dataset contains 157 records with 8 columns detailing different crime categories and their occurrences.

Crimes are categorized under "Major Heads" and "Minor Heads", with different motivations or reasons.

Crime Trends Over Time:

The dataset provides crime counts for:

The current month under review.

The previous month.

The same month in the previous year.

The cumulative count for the current year.

These allow for trend analysisâ€”detecting increases or decreases in specific crime types.

Example: Murder Cases (IPC Sec. 302/303):

Murder for gain increased from 0 in the same month last year to 2 this month.

Murder due to property disputes had 0 incidents this month, compared to 1 in the previous year.

Murder due to personal vendetta decreased from 2 last month to 1 this month.

Possible Analysis Directions:

Crime hotspots: Identify crime categories with the highest frequency.

Seasonal variations: Compare current month vs. previous month trends.

Year-over-year analysis: Identify crime categories that have significantly increased or decreased.

Predictive insights: If certain crimes spike in certain months, preventive measures can be suggested.
